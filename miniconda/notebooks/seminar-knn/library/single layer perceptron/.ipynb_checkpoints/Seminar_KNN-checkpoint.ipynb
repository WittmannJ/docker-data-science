```python
from random import choice
import matplotlib.pyplot as plt
from numpy import dot, array, random
```

```python
training_data_and = [
    (array([0,0,1]), 0),
    (array([0,1,1]), 0),
    (array([1,0,1]), 0),
    (array([1,1,1]), 1),
]
```

```python
#heaviside = lambda x: 0.5 if x==0 else 0 if x==0 else 1
def H(x):
    if x < 0:
        return 0
    elif x >= 0:
        return 1
```

```python
weights = random.rand(3)
```

```python
print(weights)
```

```python
learning_rate = 0.2
iterations = 1000
```

```python
errors = []
```

```python
for i in range(iterations):
    x, expected = choice(training_data_and)
    result = dot(weights, x)
    error = expected - H(result)
    errors.append(error)
    weights += learning_rate * error * x
    print (weights)
```

```python
print(errors)
```

```python
plt.ylim([-1,1])
plt.grid()
plt.xlabel('Iterationen')
plt.ylabel('Error')
plt.plot(errors)
plt.show()
```
